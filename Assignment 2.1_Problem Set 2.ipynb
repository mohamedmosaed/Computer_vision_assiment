 Use python to add numbers to pictures

```python
from PIL import Image,ImageFont,ImageDraw  # import the pil 
```

```python
# Open image
image = Image.open(r"C:\Users\mosae\Desktop\USD\AAI 521\download.jpg")
```

```python
display(image)
```

```python
# resize the image
basewidth = 500
wpercent = (basewidth/float(image.size[0]))
hsize = int((float(image.size[1])*float(wpercent)))
image = image.resize((basewidth,hsize), Image.ANTIALIAS)
display(image)
```

```python
wight, hight = image.size  # get the image size
text = "Mohamed Mosaed" # the text we want to add to teh picture
color = 'green' # chose the text size
fontsize = wight//20 
font = ImageFont.truetype('/root/pyscripts/arial.ttf',fontsize) # font type format
draw = ImageDraw.Draw(image) 
draw.text((fontsize*6,10), text, color, font) # show the text on the image 
display(image)
```
